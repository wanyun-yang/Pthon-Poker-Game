```python
import numpy as np
```

```python
def deal():
    # Initialize player hands
    p1 = []
    p2 = []
    
    # Initialize and shuffle deck
    deck = list(range(1, 14)) * 4
    np.random.shuffle(deck)
    
    # Deal cards to each player
    for i in range(len(deck)):
        if i % 2:
            p2.append(deck.pop(0))
        else:
            p1.append(deck.pop(0))
            
    return p1, p2
```

```python
#Play multiple games within the function, using a new argument games (equals 1 by default)
def play(games=1):
     # Collect statistics
    S = {}
    for t in range(games):
        #print('Game '+str(t+1)+' : ')
    # Deal cards to players, initialize claim lists
        p1, p2 = deal()
        p1_initial=sum(p1)
        p2_initial=sum(p2)
        p1_claim, p2_claim = [], []
        #Track the number of tie
        tieCount=0
        #Track the largest total won during a round 
        maximumTotal=0
    
    # Calculate initial statistics
    
    # Play!
        k = 1 # round counter
        i = 13 # maximum number of rounds
        while i > 0:
        # Print round
        
        # Each player plays two cards
            p1_play = [p1.pop(0), p1.pop(0)]
            p2_play = [p2.pop(0), p2.pop(0)]
            i -= 1
        
        # Check for tie, and process accordingly
            if sum(p1_play) > sum(p2_play):
                # Player 1 wins hand, append cards to claim list
                p1_claim.extend(p1_play); p1_claim.extend(p2_play)
                maximumTotal=max(maximumTotal,sum(p1_play))
                #print('\tPlayer 1 wins hand (%d over %d), collects' % (sum(p1_play), sum(p2_play)), sum(p1_play) + sum(p2_play))
            elif sum(p2_play) > sum(p1_play):
                # Player 2 wins hand, append cards
                p2_claim.extend(p1_play); p2_claim.extend(p2_play)
                maximumTotal=max(maximumTotal,sum(p2_play))
                #print('\tPlayer 2 wins hand (%d over %d), collects' % (sum(p2_play), sum(p1_play)), sum(p1_play) + sum(p2_play))
            else:
                tieCount+=1
                # Tie - Continue to play two cards until a player wins
                #print('\tTie:', sum(p1_play), 'vs.', sum(p2_play))
                if i > 0:
                    while sum(p1_play[-2:]) == sum(p2_play[-2:]):
                        p1_play.extend([p1.pop(0), p1.pop(0)])
                        p2_play.extend([p2.pop(0), p2.pop(0)])
                        i -= 1
                        if sum(p1_play[-2:]) > sum(p2_play[-2:]):
                            # Player 1 wins hand, append cards to claim list
                            p1_claim.extend(p1_play); p1_claim.extend(p2_play)
                            maximumTotal=max(maximumTotal,sum(p1_play))
                            #print('\tPlayer 1 wins hand (%d over %d), collects' % (sum(p1_play[-2:]), sum(p2_play[-2:])), sum(p1_play) + sum(p2_play))
                        elif sum(p2_play[-2:]) > sum(p1_play[-2:]):
                            # Player 2 wins hand, append cards to claim list
                            p2_claim.extend(p1_play); p2_claim.extend(p2_play)
                            maximumTotal=max(maximumTotal,sum(p2_play))
                            #print('\tPlayer 2 wins hand (%d over %d), collects' % (sum(p2_play[-2:]), sum(p1_play[-2:])), sum(p1_play) + sum(p2_play))
                        else:
                            tieCount+=1
                            maximumTotal=max(maximumTotal,sum(p1_play))
                            #print('\t\tAnother Tie:', sum(p1_play[-2:]), 'vs.', sum(p2_play[-2:]))

                        # Check for whether game ends in a tie - return cards to respective players
                            if i <= 0:
                                p1_claim.extend(p1_play)
                                p2_claim.extend(p2_play)
                                break
                else:
                    # Game ends in a tie - return cards to respective players
                    p1_claim.extend(p1_play)
                    p2_claim.extend(p2_play)
                    
            # Increment round
            k += 1
    
        # Game over, compile results
        S['Game ' + str(t+1)] = [p1_initial,p2_initial,tieCount,maximumTotal,sum(p1_claim),sum(p2_claim)]
        # Return results
    return S
```

```python
#Demonstrate the function for 1 game
print('Each game: Player 1 initial total, Player 2 initial total, Number of ties during gameplay, Maximum round total won, Player 1 final total, Player 2 final total are below by order:' )    
play(1)
```

```python
# Calculate the mean maximum round total as a function of the number of ties
def maximumTotal(tie=0):
    return sum(ThousandGames[ThousandGames['# of tie']==tie]['Maximum total in the round'])/len(ThousandGames[ThousandGames['# of tie']==tie])
```

```python
# The biggest possible round number is 13 rounds.
# Calculate the frequency of the number of ties within each game
tieList=[]
meanMaximumList=[]
for i in range(14):
    if len(ThousandGames[ThousandGames['# of tie']==i])!=0:
        tieList.append(i)
        meanMaximumList.append(maximumTotal(i))
        print('For tie number '+str(i)+': There were '+ str(len(ThousandGames[ThousandGames['# of tie']==i]))+' games. The mean maximum total is ' + str(maximumTotal(i)))
```

```python
tieCorr=np.corrcoef(tieList, meanMaximumList)[0, 1]
tieCorr
print('The number of ties has a positive relation with mean maximum total')
```
